```python
#Construct the Phylogeny
!augur tree --alignment Practical1_Data/aligned.fasta --output Practical1_Data/tree.nwk
```

```python
#Infer Ancestral Sequences
!augur ancestral \
  --tree Practical1_Data/tree.nwk \
  --alignment Practical1_Data/aligned.fasta \
  --output-node-data Practical1_Data/nt_muts.json \
  --inference joint
```

```python
#Identify Amino-Acid Mutations
!!augur translate \
  --tree Practical1_Data/tree.nwk \
  --ancestral-sequences Practical1_Data/nt_muts.json \
  --reference-sequence Practical1_Data/emu.fas
  --output-node-data Practical1_Data/aa_muts.json
```
